```python
plt.figure(figsize=(10,5))
plt.bar(Q1.index, Q1.values, color='skyblue')
plt.title("Count of country")
plt.xlabel("Country")
plt.ylabel("Count of Country")
plt.show()
```

```python
Q2 = aadi['Year'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(Q2.values, labels=Q2.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Years')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(aadi['Literacy Rate (%)'], aadi['Poverty Rate (%)'], color='yellow')
plt.title('Literacy Rate vs Poverty Rate')
plt.xlabel('Literacy rate')
plt.ylabel('Poverty Rate')
plt.grid(True)
plt.show()
```

```python
Q4 = aadi.groupby("Country").agg({"Number of Airports" : "max"})
Q4
```

```python
plt.figure(figsize=(8,5))
plt.bar(Q4.index, Q4.values, color='lightcoral')
plt.title('Number of Airports per Country')
plt.xlabel('NUmber of airports')
plt.ylabel('Country')
plt.show()
```
